```julia
using RegressionDynamicCausalModeling
using Plots
using Random
```

```julia
dcm = load_DCM("../SRPBS_example/sub-0001/DCM.mat")
```

```julia
heatmap(
    dcm.a;
    yflip=true,
    title="Network architecture (A matrix)",
    titlefontsize=8,
    legend=:none,
    xlabel="region from",
    ylabel="region to",
    aspect_ratio=1,
    size=(400, 400),
)
```

```julia
reg = 1
p1 = plot(dcm.Y.y[:, reg]; label="$(dcm.Y.name[reg])", title="BOLD signal of region $(reg)")
```

```julia
dcm.c # no driving inputs
```

```julia
opt = Options(RigidInversionParams(); synthetic=false, rng=MersenneTwister(42))

rdcm = RigidRdcm(dcm) # convert the DCM to a rDCM model
```

```julia
# Estimate effective connectivity
output = invert(rdcm, opt)
```

```julia
# plot the posterior mean of the A matrix (in output.m_all)
heatmap(
    output.m_all[:,1:end-1];
    yflip=true,
    title="Posterior mean (A matrix)",
    titlefontsize=8,
    c=:viridis,
    xlabel="region from",
    ylabel="region to",
    aspect_ratio=1,
    size=(400, 400),
)
```
